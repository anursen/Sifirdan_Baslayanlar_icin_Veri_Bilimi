2. Basamak: Veriyi Anlamak.
* Orta ve büyük ölçekli şirketlerde veri, veri tabanlarından gelmektedir.

image.png

* Şuan da nihai olarak bir model oluşturacağız.
* Bu model bir Makine Öğrenmesi modeli olacaktır.
* Bu model aracın özellikleri girildiğinde bize aracın fiyatını tahmin edecek yapay öğrenilmiş bir sistem ortaya çıkarıcaktır.
* Veri kaynaklarından veriyi aldığımızda bu veriyi anlamamız, incelememiz gerekiyor.


image.png

* Sayısal değişkenlerin ortalaması, standart sapması gibi betimsel istatistiklerini bakmalıyız.
* Marka model dağılımında örneğin A markasına ait A1 modellerin fiyat ortalaması nedir, A markasının manuel viteslerinin ortalama fiyatları nedir otomatik vitesli olanların ortalama fiyatları nedir gibi veriyi anlayacağımız işlemler gerçekleştireceğiz.
* Sonuçta insan olarak veri tabanımızda yer alan ve excel formunda olan  bu veriyi anlamamız gerekiyor.
* Veriyi değişkenlerini, değişkenlerin dağılımlarını, değişkenlerini tiplerini, ortalmasını, vb. şeklinde tanımamız gerekiyor.
