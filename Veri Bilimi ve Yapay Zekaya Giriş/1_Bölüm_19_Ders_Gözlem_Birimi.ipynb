**Gözlem Birimi nedir?**
* Araştırma da incelediğimiz birimlerdir.

image.png

* Önceki derste ana kitle içinden örneklem seçtik.
* Örneklemin içindeki her bir eleman gözlem birimi olarak adlandırılır.

image.png

* Seçim çalışmasını düşünecek olursak mikrofon uzatılan hangi partiye oy verirsiniz sorusunuz cevabını aldığımız her bir bireydir gözlem birimi.
* Ya da bir ML çalışması yapıldığında incelenen gözlem birimi farklılaşacaktır.

image.png

* Araç fiyat tahmin modelindeki veri setinde yer alan herbir satır bir gözlem birimidir.
