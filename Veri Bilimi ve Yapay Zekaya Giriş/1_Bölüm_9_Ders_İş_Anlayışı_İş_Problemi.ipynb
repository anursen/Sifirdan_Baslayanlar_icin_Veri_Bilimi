* Bu bölüm de bir iş sektörünün içerisinde bir iş süreci döngüsünün sürecinde nasıl gerçekleştirebileceğimiz görmüş olacağız.
* İş dünyasında Data Mining veya Data Science süreçlerinin modellenmesi gibi bir durum söz konusu.
* Bu akışı modelleyen yaklaşımlardan birisi CRISP-DM metodolojisi.
* İfade edildiği üzere Veri Bilimi süreçlerinin iş dünyasındaki  ele alınışını ifade ediyor.
* Sıfırdan bir ürün geliştirme işlemini şirketler genelde bu dögüyü takip ederek gerçekleştiriyor.
* CRISP-DM metodolojisi üzerinden bir iş problemini bir iş anlayışını veriyi anlamak basamağını, veri tabanlarında bulunan veriyi anlamayı verinin hazırlanmasını modelleme işlemini (bu kısım ML kımıdır), oluşturulan modelin başarısının değerlendirilmesini ve insanların kullanımına sokulması gibi basamaklar ile bu bölümü ele almış olacağız.

* Öncelikle bir problem ortaya çıkaracağız ve bu problem üzerinden gerçek hayatta bir veri bilimi sürecinin nasıl ele alındığını değerlendirmiş olacağız.
* **1. Basamak:** 
* İş Anlayışı.
* Yani iş problemidir.
* E-ticaret sektöründe ikinci el ilan sitelerini ele alacağız ve iş anlayışıyla bir problemi projeye başlamış olacağız.

* Bir otomobilimiz var ve bunu satmak istiyoruz.
* Buna yönelik olarak ikinci el ilan sitesine giriyoruz ve aracımızı satmak için fiyat belirlememiz gerekiyor.
* Fakat arabalardan pek anlamadığımız için bizim arabamızın fiyatının ne olabileceğini aracımıza benzer binlerce ilanı gezip aracımızın fiyatını piyasada yaklaşık olarak ne şekilde konumlandığını araştırmakla uğraşmak istemiyoruz.
* Aracını satmak isteyen bir kişi aracını ne kadar satabilirim acaba diye bu şekilde bir problemle karşılaştığında karşılaştıkları bu problemleri ikinci el ilan sitesine bildirdikleri durumda bir iş sektörü içerisinde bir iş ihtiyacının ortaya çıkması söz konusu oluyor.

image.png

* Ortaya çıkan bir iş problemi çözülmek üzere bir Veri Bilimi takımına gelmiş oluyor.
* **Problem:**
* Araçların fiyatlarını daha kolay bir şekilde belirlemek istiyorlar gerçeğe yakın bir şekilde belirleyebilmek istiyorlar.
* Kullanıcı olarak yapmak istediğimiz şey şu:
* Aracın km, yaş, kaza durumunu, vites durumunu belirttiğimizde bize ilgili arabanın benzer marka model vb. durumlar göz önünde bulundurularak kaç para edebileceğini bilgisi verilsin 

image.png

* Örneğin bir galerici olduğunu düşünelim ve bu galerici araç önüne geldiğinde araç hakkında birkaç soru sorar ve sonuç olarak piyasayıda bildiğinden dolayı fiyat verebilir.
* Ama biz b,rey olarak uzman galerici gibi piyasayı bilemediğimizden dolayı bu arabaya bu şekilde bir fiyat veremiyoruz. 
* Yetersiz kalıyoruz bu da bizim için problem demek.


image.png

* Bu problem Veri Bilimcinin önüne geldi diyelim, peki o nereden bilecek bunun fiyatını.
* Bu duurmda elindeki verilere bakacak tabi ki.
* Galericinin edinmiş olduğu deneyimdeki gibi veri setlerinde hangi marka modelin kaç km'de kaza durumu olup olamamsı durumunda, hasar durumu olup olmaması durumunda hangi fiyatlarla listelendiğini ilan sitesinde (veri tabanlarında) veriliyor.
* Problemin çözümünde nihai olarak yapacak olduğumuz şey ana amacımız burada bir model oluşturmak, bir ML modeli oluştırmaktır.
* Bu model ne yapacak?


image.png

* Bu modeli bir fonksiyon gibi düşüneceğiz.
* Araçlara ilişkin değerler bu fonksiyona yazıldığında aracın ffiyatını bize tahmin ediyor olacaktır.
* Veri kaynaklarına gideceğiz ve adım adım bu işlemi gerçekleştirmeye çalışacağız.
* Biz bir Veri Bilimci olarak aracın fiyatını galerici gibi bilemeyiz.
* Ama veri tabanlarımızda yüz binlerce aracın hangi özellikler sergilendiğinde, hangi özelliklere sahip olduğunda hangi fiyatlarla satıldığınıbiliyoruz.
* Biz algoritmaya deneyimli bir galerici gibi verilerle öğretmiş olacağız.
