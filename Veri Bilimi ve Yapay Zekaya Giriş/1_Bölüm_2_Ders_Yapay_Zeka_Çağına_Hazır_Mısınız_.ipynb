* Tüm dünyanın bu yapay zeka çağına olan bakış açısını şekillendiren iki önemli cümle:

image.png

* Dijitalleşen dünya da tıpkı petrol gibi işlenmeyi hazır hale getirilmeyi ve bazı değerler üretilmeyi bekleyen şey artık veridir.
* Bir diğer söz ise Stanford Üniversitesi'nde profesör, Coursera eğitim platformunun da kurucusu olan Andrew Ng'ye aittir.
* Andrew Ng'nin söylediği bu sözden bakış açısı ve vizyonu ile Yapay Zekanın da günümüzün elektriği olduğu ve hayatımızın her alanında kullanılacak olduğunu ifade etmektedir.
* Veriden beslenen Yapay Zeka günümüzün yeni elektriği olarak tanımlanmaktadır.
* **Yapay Zekanın Tanımı:**

image.png

* Farkında olsak da olmasak da günlük hayatımızda Yapay Zeka uygulamaları ile etkileşime girmiş durumdayız.
* Yapay Zeka çalışmalarını inceleyelim.

image.png

* Sosyal medya hesaplarında arkadaş önerileri "Tanıyor Olabileceğiniz Kişiler".

image.png

* Yüklenen bir fotoğrafın isim etiketlemeleri otomatik olarak yapılması.
* Facebook uygulamasında sıkça görülen bir durumdur.

image.png

* Ve daha ileri seviyede yüklenen fotoğrafların etiketlemeleri çeşitlenmiş durumda. 
* Yani fotoğraflarımız içeriklerine göre sınıflandırılıyor kategorize edilip öylecek saklanıyor.

image.png

* Sosyal medyada ilgilendiğimiz içeriklerin belirlenip sürekli benzer içeriklerin karşımıza gelmesi.
* Sosyal medya kullanımı esnasında bir örnek video izlediğimizde veya Youtube platformunda müzik dinlediğimizde öneri algoritmaları bizim izleme geçmişlerimize en yakın olduğu şekilde bazı kümelemeler yaparak bize bu ortamlarda ilgilenebieceğimiz içerikleri sürekli karşımıza getirmeye çalışıyor. (İsmi kişiselliştirme kişiye özel olarak geçiyor.)

image.png

* Yazdığımız mesajların otomatik olarak tamamlanması.
* Merhaba kelimesinin ilk birkaç harfini yazdıktan sonra bize kelimeyi öneriyor.

image.png

* Araştırma yaptığımız ürünün ya da yakınımızda biriyle bir ürün hakkında konuşmamız sonucunda (eğer uygulamaların mikrofon özelliği izni verilmişse) o ürün vb. ürünlerin sürekli internet sitelerinde karşımıza çıkması. 
* Buna da hedefli ürün pazarlama denir.
* Sosyal medya tarafından verilen içerikleri tüketiyoruz.
* Sosyal medya tarafından verilen bu içerikler açısından bir kaygı da bu içeriklerin tüketilmesidir.
* Bize öyle içerikler sunulmalı ki o içerik ilgimizi çeksin.
* Buna da hedefli içerik denir.
* Benzer şekilde bize sosyal medya ve internet dünyasında öyle ürünler sunulmalı ki bu ürünler bizim ilgilenebiliyor olacağımız ya da ilgilendiğimiz ürünler olsun.

image.png

* Bir online alışveri sitesinde bir ürün incelemesi yaptığımız sırada o ürünü alan aşağıdaki ürünleri de aldı gibi bildiirmlerle karşılaşıyoruz.
* Bu ve buna benzer uygulamalar günlük hayatımızın ilk temas noktalarında karşılaştığımız uygulamalardır.
