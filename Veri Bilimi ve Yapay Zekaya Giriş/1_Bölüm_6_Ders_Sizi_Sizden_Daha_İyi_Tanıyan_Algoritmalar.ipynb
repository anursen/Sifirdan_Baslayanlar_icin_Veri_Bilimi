* Algoritma nedir?
* Genel anlamıyla belirli bir amacı yeri getirmek için gerçekleştirilen işlemler dizisidir.
* Belirli bir amaçları yerine getiren algoritmaları eğer veriyle besler ve onlara öğrenme kabiliyeti kazandırabilirsek (ML, DL, İstatistiksel Öğrenme, Yapay Zeka olarak bunları kullanabilirsek) bu durumda bizi bizden daha iyi tanır hale gelebilirler.
* Görüldüğü üzere sizi sizden daha iyi tanır algoritmalar söz konusu olacak.
* Tanımlayamadığımız kişisel eğilimlerimiz, meraklarımız, hassas noktalarımız bu algoritmalar tarafından tanımlanabilir bir hale geliyor.


image.png

* Mikhail Kozinski ortaya attığı bu algoritmik yaklaşım Cambridge Analytica şirketi tarafından uygulanarak ABD'nin seçim çalışmalarında bu basit analitik düşünce genellenip ölçeklendirlerek sosyal medya üzerinden ABD seçmeninin 17 eyalet 32 farklı kişilik grubuna ayırıyorlar.

* Gruplara ayırma işlemine persona denir.
* Persona özelinde kişiselleştirilmiş seçim kampanyaları oluşturarak ABD başkanının seçilmesinde analitik düşünce becerilerinin rolünü ortaya koyuyorlar.

image.png

* Trump tüm seçmenlerine dinamik bir içerik stratejisi oluşturuyor.
* Örneğin sınırdaki geçişler konusunda hassasiyeti olan bir kitleye onları rahatlatacak bir açıklama yaparken genç gruplara farklı iletişim stratejileri izliyor.
* Burda veri analitği, analitik bakış açısı bir devletin ve dolayısıyla dünyanın geleceğinde rol oynuyor.
* Devletlerin milletlerin ve şirketlerin bu yapay zeka çağında hayatta kalması veri analitiği ve analitik düşünce beceileriyle mümkün olacaktır.
* Burada ortaya çıkan değere Veri Bilimi ve Veri Analitiği denir.
* Uçtan uca bu süreci ele aldığımızda Veri Bilimi süreci vb. işleri yapan kişilere ise Veri Bilimci denir.
