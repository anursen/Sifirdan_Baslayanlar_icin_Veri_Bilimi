* Elimizdeki veriyi ne şekilde nasıl sunmalıyız?

image.png

**A şıkkı:**
* Grafikler ev fiyalarındaki artışı ortaya koymak için kullanılır.
* Ev fiyatları artış konusunda bilgi vermektedir.

**B Şıkkı**
* İki grafik karşılaştırıldıüğında bir problem olduğu görülmekte.
* Soldaki grafiğin y ekseni 79K'dan başlamış, sağdaki grafik ise y ekseninin başlangıç noktası 0K'dır.
* Grafik gösterme tekniklerinde eksenlere konulacak olan değişkenin ölçeği eldeki verinin sunumunda suistimale en açık olan konulardan birisidir.
* Eksenlerin başlangıç noktaları sıfır olmak durumundadır.
* Eksenin başlangıç noktası açısından sağdaki grafik daha doğrudur.
* Sağdaki grafiğin ölçeği 10K (10.000'er) şeklinde artarken soldaki grafiğin ölçeği ise 1K (1'er) şeklinde armış.
* Soldaki grafikteki gibi ev fiyatlarını 1k 1k artış ile gösterirsek bu durumda fiyat artışları grafikte de görüldüğü üzere çok yüksek gözükecektir.
* Ev fiyatlarını haberleştirmeyi düşünüldüğünde 1000TL gibi bir artışın bu kadar yüksek ölçekte gösterilmesi ev fiyatının lendi içindeki yayılımı göz önünde bulundurulduğunda çok mantıklı olmayacaktır.
* Sağdaki grafik ve yorum seçilir.
