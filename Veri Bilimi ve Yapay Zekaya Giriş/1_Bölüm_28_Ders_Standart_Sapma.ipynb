* Herbir gözlem biriminin ortalama değere olan uzaklıkları aldığımızda ortalamaya olan uzaklığı hesaplamış oluyoruz.
* Ortalamadan olan bu sapmaların genel bir ölçüsü ise n'e böldüğümüzde sapmaların ortalamasını hesaplamış oluyoruz.
* Standart sapma ortalamalara olan bir sapmanın ortalamasıdr.

image.png

* Yani x ortalama ile mü aynı şeylerdir.
* Biri ana kitleyi diğeri ise ana kitleyi temsil eder.
* Grafikte görüldüğü üzere ana kitle ortalamasına + - yönlü  bir standart sapma mesafe konulmuş.
* Bu grafik bize ortalama etrafındaki dağılımını ifade ediyor.
* Grafikte standart sapma ifadesi sigma ifadesi verilen ifade ile gösterilmiş.
