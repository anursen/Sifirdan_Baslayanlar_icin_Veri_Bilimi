# **18.Ders: Popülasyon ve Örneklem**

image.png

image.png

* Temel Kavramları çğreniyorum.
* **Popülasyon Nedir?**
* İlgilendiğimiz ana kitledir.

* **Örneklem Nedir?**
* Ana kitle içinden seçilen ana kitleyi temsil niteliği yüksek bir alt kümedir.
* Bizler veriden faydalı bilgiler çıkarmaya çalışıyoruz. 
* Bazen verinin hepsi bazen de ana kümeyi temsil eden örneklem olur.
* Gerçek hayat uygulamalarında elimizde bütün veri olmaz da onu temsil eden örneklem olur.
* Verinin tamamına popülasyon, o veriyi temsil eden alt kümeye örneklem denir.
* **Dikkat:** Örneklem ana kitleyi çok iyi temsil etmesi gerekir.
* Çünkü örneklem san ki popülasyonmuşcasına muamele görecektir.
* Dolayısıyla elde ettiğimiz örneklemin popülasyonun niteliklerini temsil etmesi ve yansız olması gerekir.
* Popülasyon içinden örneklem çekmek için bazı yöntemler var.
* Örnek: Seçim zamanlarında A, B ve C partilerinin seçim anketleri açıklanıyor.
* Bu anketlerin bütün ülkeye yapılmıyor.
* Yani ülkeyi popülasyon olarak kabul edersek içerisinden iyi bir temsilici olduğu varsayılan örneklem çekiliyor.
* Türkiye için bu örneklem rakamı yaklaşık olarak 2500'dür.
* Tabi ki bu örneklem çekme de belirli bir hata göze alınır.
* Mesela %95 güvenilirlik ile veya %99'luk güvenilirlik ile oy oranları şu şekilde elde ediliyor şeklinde çıkarımda bulunuluyor.
* %95 gibi %99 gibi belirli bir güven aralığı vererek istatistiki sapmayı da (%5, %1) göz önünde bulundurarak bütün kitleyi incelemek yerine bu kitleyi temsil eden örnek üzerinden çalışmalar gerçekleştirmiş olunuyor.
