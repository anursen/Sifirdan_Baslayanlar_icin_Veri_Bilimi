* Dayanıklılık skorunun yüksek olması iyi birşeydir.
* Standart Sapma değerini görmeden seçme işlemi yapmak istersek o halde doğal olarak ortalaması yüksek olan değeri yani ünlüleri seçmemiz gerekirdi.
* Bir standart sapma basit bir dağıolımdan çok daha fazla bilgi taşımaktadır.
* Ünlülerin Standart Sapma değeri çok yüksek.
* Bu demek oluyor ki (ünlüler için) 80'den daha fazla ve 66 civarında kişilerde olabilir.
* Gönüllüer seçilir.
* Çünkü yetenek açısından birbirlerine daha yakınlar (standart sapması düşük) ve daha homojenler birbirlerine benziyorlar.
* Gerçek hayatta iş süreçleri anlamında sayılar rakamlar anlamında bir yerlere entegre ettiğimiz de geldiği anlamlar bunlar olacaktır.
* Standart Sapmanın yüksek olması (örneğin aylara göre satış ort. değerleri incelendiğinde) böyle bir durumda standart sapmanın yüksek olması ayların birbirlerine göre çok değişkenlik gösterdiği anlamına gelirken standart sapmanın düşük olması satışların birbirlerine çok benzer olduğu anlamına gelir.
