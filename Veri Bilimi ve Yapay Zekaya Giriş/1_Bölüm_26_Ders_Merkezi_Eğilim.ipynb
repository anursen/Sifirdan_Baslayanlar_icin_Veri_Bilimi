2. **Doğru Kullanılması**

**Örnek**

image.png

* İki emlakçı bir konuda anlaşmazlığa düşümüşler.
* Anlaşamadıkları konu hangisinin portföyü daha iyi olduğudur.
* Yani satışa hazır ev fiyatları.
* İki emlakçıda kendi kafasında bir bedel belirlemişler ve kendilerinin daha yüksek segmentlere hitap ettiklerini düşünüyorlar

image.png

* Böyle bir soru sormuş olalım emlakçılara.
* **Not:** Ortalama ile Medyanın birbirine yakın olması veri setinin düzgün dağıldığını, homojen olduğunu gösterir.

image.png

* Günlük hayatta bazı kavramalrı nümerik olarak bazı değerlere dönüştürebilmek bir meziyet.
* Fakat bir de bunları doğru kullanabilmek gerekiyor.

* A Emlakçı iddasında daha doğrudur.
* B Emlakçıda yer alan 3.500.000TL değerindeki evdir.
* Bu değer aykırı değerdir. 
* Serinin geel yapısından uzakda duran bir değerdir.
* Ortalamayı yukarı çeken değerdir bu değer.
* Bundan dolayı yani B emlağı simetrik olmadığı için iki değişkeni de kıyaslamak için en doğru metrik medyan değeridir.
* Kendi içinde kıyaslayacak için A Emlağı için ortalmayı da Medyan değerinide kullanabiliriz.
* B Emlağını değerlendireceksek ortalama değeri kullanamayız.
