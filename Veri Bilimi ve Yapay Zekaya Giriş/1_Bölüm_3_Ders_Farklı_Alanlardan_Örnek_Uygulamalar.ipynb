* Örneğin bir hastanın MR görüntülerinin incelenerek kanser hastalığına sahip olup olmadığını tahmin edilmesidir.
* Google'ında içinde bulunduğu akciğer kanseri teşhisi çalışmasında Yapay Zeka yaklaşık %94 başarı ile kanseri teşhis edebiliyor.
* Peki Yapay Zekanın bir hastalığı bu kadar yüksek başarıyla teşhis edebilmesinin arkasındaki mantık nedir ve bu bizim için neden önemlidir?
* Göreve yeni başlayan bir doktor düşünelim ve 20 30 yıldır işin içinde olan bir doktoru düşünelim.
* Birinci profildeki göreve yeni başlayan doktor MR görüntülerini inceleyerek bir akciğer kanserine karar verme noktasında çok başarılı olamayabilecektir. 
* Çünkü tecrübesizdir, göreve yeni başlamıştır vs.
* Göreve yeni başlayan bu doktorun çok başarılı bir şekilde kanser hastalığına ilişkin MR görüntüleri üzerinden bir sonuca varabilmesi için ikinci profildeki doktor kadar deneyim kazanması gerekir.
* Daha fazla hasta, kanserli ve kanserli olmayan MR görüntüsü görmesi gerekir.
* Bunların neticesinde edindiği deneyimle belirli bir süre sonra bir görüntüye baktığında bu hasta kanser hastalığına sahiptir veya değildir şeklinde teşhis edebilir bir seviyeye gelecektir.
* Yapay Zekanın önemi şurada devreye giriyor:
* Bir doktorun 20-30 yılda kazanabileceği deneyimi biz bir sisteme entegre edip ona kanserli olan görüntüleri ve kanserli olmayan görüntüleri verip onu eğittiğimizde Yapay Zeka sistemimiz burada kanserli görüntüler ile kanserli olmayan görüntüler arasındaki ayrıştrıcılığı tespit edip bu kişi kanserli veya değildir yorumu yapabilir seviyeye gelecektir.

* Şirketlerin yıl sonu satış ve gelirlerini şimdiden tahmin edebiliyor olması ve bütün departmanlara buna göre görevler ve hedefler verebiliyor olmasıdır.
* Bir şirket düşünün ki satışlarını gelirlerini, kendisi terk edebilecek müşterilerini hatta kendisi terk edebilecek çalışanlarını dahi tahmin edebiliyor ve buna göre hareket edebilir duruma geliyor.

* Bir kredi başvurusunda bulunduğumuzu düşünelim ve bu kredinin ödeyip ödeyemeyeceğimizin merak edildiğini düşünelim.
* Bu krediyi ödeyip ödeyemeyeceğini tahmin etmek işleri örnek uygulamalar arasındadır.
* Peki bu iş nasıl gerçekleşiyor?
* Veri tabanı olarak adlandırılan verilerimizin tutulduğu yerlerdeki verileri inceleyerek bu veriler üzerinden ilgilendiğimiz konuya göre bazı yapıları algoritmalar aracılığı ile öğreniyoruz.
* Bir bankacıya bir kişinin kredi başvurusu yaptığını düşünelim.
* Yeni işe başlayan bir bankacı olarak bu kişinin kredisini ödeyip ödeyemeyeceğini bilemeyebiliriz.
* Basit olarak varlıklarına, gelir durumuna bakarak karar verebiliriz.
* Buna benzer bir şekilde bir sistem düşünelim. 
* Bu sistem geçmişteki yüz binlerce kişinin verilerine bakarak bir karara varmaya çalışıyor.
* Krediyi ödeyenlerin ve ödeyemeyenlerin birbirinden nasıl ayrıldığının yapısını öğrenmeye çalışıyor.
* Sonuç olarak bir kullanıcı Yapay Zeka sistemimize geldiğinde bu kullanıcının gelir, çocuk sayısı, geçmişteki kredi ödemelerine ilişkin performans ölçümleri vb. değişken değerlerini verdiğimizde bu sistem geçmiş verilere bakarak bir yapı öğrenip bu kişinin krediyi ödeyip ödeyemeyeceğine karar verir.
