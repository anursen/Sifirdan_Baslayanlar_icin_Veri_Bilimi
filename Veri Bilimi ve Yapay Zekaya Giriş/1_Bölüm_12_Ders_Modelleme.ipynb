**Modelleme:**
* Veri içerisinde yer alan yapıların algıritmalara öğretilmesi demektir.
* Veri seti içindeki bu yapıyı birçok algoritma ile öğrenip modelleyebiliriz.
* Öğrenme türlerinden birisi Çoklu Regresyon modelidir.

image.png

* Burada aracın bilgilerini girdiğimizde fiyatını değimiz şey görselde görülen fonksiyondan ibaret olacaktır.
* b0, b1, .. bp değerleri veri setinin içerisinden öğrenilen katsayılardır.
* Veri setinin içindeki yapıyı öğrenilmesi demek aslında fiyat söz konusu olduğunda fiyatı etkilediğini bildiğimiz km, hasar durumu vb. değişkenlerin etkilerinin yönünü ve şiddetini bulmak demektir.


image.png

* Bu fonksiyon benzeri bir yapı bulduğumuzda değişkenlerin katkılarını (etkilerini) bulmuş olacağız.
* Mesela aracın vites durumu fiyatına direkt etki eder.
* Ya da kazalı olması, km değerinin yüksek olması fiyata negatif etki eder.
* Ama örneğin A marka B model aracın km değerinin arttığında fiyata nasıl arttığını bilmiyoruz.
* Bu artış miktarını sektörden biri olan galericilerde bilmiyor.
* Mesela aracın yaşlı olması fiyatını negatif etkiler ama km değeri düşükse (iki faktörü bir arada göz önünde bulundurmaya çalışıyoruz) fiyatına olumlu etki eder.


image.png

* Netice itibariyle veri setine çoklu doğrusal regresyon modelini verdiğimizde değişkenler arasındaki lişkiyi modellediğimizde bir fonksiyon ortaya çıkacaktır.
* Beta değerlerin (b0, b1, .... bp) veri setininin içerisinden öğrenecek olduğumuz değişkenlerin etki düzeylerini yönlerini belirtecek olan katsayılar olduğunu ifade etmiştik.
* Nihai olarak Tahmini Araç Fiyatı olarak bir fonksiyon oluştu elimizde.
* b0, b1, b2, ... bp ifadeleri veri setinin içerisinden öğrenmiş olduğumuz katsayılar (b0 hariç tutarak çünkü o sabit değer, b1 den itibaren ağırlıklar yani w değerleridir.)
* Çoklu Doğrusal regresyon algoritması veri içerisinde fiyatı etkileyen faktörleri yönünü ve şiddetini bulacaktır.
* Mesela hasar durumu değişkenini katsayısı negatif yönlü şiddeti ise 2500.
* Araçta eğer hasar varsa fiyatını 2500TL aşağıya çekecektir.
* Vites türü değişkeni 1 (yani Otomatik vites olmasının encode edilmiş halidir) yönü pozitif, şiddeti ise 6000'dir.
