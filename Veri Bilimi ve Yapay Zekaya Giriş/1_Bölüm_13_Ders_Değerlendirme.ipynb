# **13.Ders: Değerlendirme**

* Derğerlendirme bölümündeyiz.
* Değerlendirme ne demek?
* **Dikkat:** 
  * Her zaman Veri Bilimi projelerinde bir Makine Öğrenmesi projesi kurmak gerekmez.
  * Bir istatistiksel test, rapor veya simülasyon çalışması da yapılabilir.
* Yapılan bu projenin bir değerlendirmesi yapılır.
* Karşılaştırma, performans değerlendirme, testler yapılması gerekir.
* Tahmin modelimizin performansını değerlendireceğiz.
* Oluşturduğumuz tahmin model fonksiyonunu test etmek istiyoruz.
* Modele çıktısının ne olduğunu bildiğimiz bir veri veriririz ve çıktıyı tahmin etmesini bekleriz.
* Sonuç olarak gerçek değerler ile tahmin edilen bu değerleri karşılaştırma yapıyoruz.

image.png

* Veri setinin içinde bu şekilde bir gözlem olduğunu düşünelim.

image.png

* Tahmin fonksiyonumuzu da getirelim.

image.png

* Yukarıdaki gözlem değerlerini fonksiyona yazdığımız zaman 71500TL olarak aracın fiyatını tahmin etti.
* Gerçek değerimiz ise 73000TL.
* Bu değerler arasındaki farkları değerlendirerek bir model değerlendirme işlemi gerçekleştiririz.
* Biz şuanda tek bir örnekte inceledik.
* Yani Tahmin Edilen Fİyat ile Gerçek Fiyat arasındaki fark sadece tek bir gözlem üzerinden elde ettiğimiz hatadır.
* Bu işlemi bütün veri seti için ya da belirledğimiz kısmı için yaptığımız da sonrasında sonrasında karelerini alıp çıkan sonuncun ortalamasını aldığımızda Hata Kareler Ortalaması (MSE)'yi bulmuş oluruz.
* SOnuç olarak MSE değeri ile bizim modelimiz şu kadar başarılıdır şeklinde bir değerlendirme de bulunuyoruz.
* Bizim amacımız bu gerçek değerler ile tahmin edilen değerler arasındaki farkları min. yapacak şekilde algoritmaları çalıştırmak olacaktır.
* Bu değerlendirme basamağı çok değerlidir ve diğer bölümlerle etkileşimlidir.
* Başarıyı değerlendirdikten sonra Veri Manipülasyonu bölümüne geri dönüp yeni değişkenler ekleyebiliriz, veri ön işleme basamaklarındaki problem giderme işlemlerini çeşitlendirebiliriz vb teknikler ile bulduğumuz bu hatayı küçültmeye azaltmaya çalışırız.
