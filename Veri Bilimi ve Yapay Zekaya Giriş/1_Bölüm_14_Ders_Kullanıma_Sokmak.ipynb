* Son bölüm Kullanıma Sokma
* Model başarı değerlendirme işleminden sonra kurduğumuz bu modeli web sitemize entegre edebilir kullanıcılara açabiliriz.
* Kullanıcılar araçlarının fiyatlarını direk özelliklerini girerek yaklaşık olarak tahmin edebileceklerdir.
* Kullanıma sokma basamağı karar verilen şekliyle (bu örnekteki gibi canlı sistemlere entegresyon şeklinde), bibir excel tablosu şeklinde bazende MySQL şeklinde gerçekleşebiliyor.
* Ya da bir rapor da oluşturulabiliyor.
* Yani satışlardaki bazı durumların neden ortaya çıktığını bulguladığımız birr raporuda sunduğumuzda bu da bir kullanıma sokmak anlamına gelir.
* Bu projeyi biz Veri Bilimciler olarak ele aldığımızda bu projeye Veri Bİlimi projesi diyoruz.
* İş olarak düşünüldüğünde ise Makine Öğrenmesi işi olarak ele alıyoruz.
* Çünkü burada tahmin eden modelin performansı ve bunun iyileşirmesi üzerinde konuşuyor oluyoruz.
* Bu projeyi şirkette üst yöneticiler ele aldığında ise buna Yapay Zeka demiş oluyor.
* Tipik bir Veri Bİlimi projesi süreci ort. altı ay sürer.
* Bu projenin en kritik noktası (modelleme ve bunun bir başarısı olduğu için) kendi takım arkadaşlarımızla değerlendireceğimiz şey Makine Mğrenmesi kısmı olacaktır.
* Bu projenin en değerli olan kısmı veri odaklı bir problemi gelire çevirmek anlamında iş odaklı düşünmektir.
* Ürüne canlı sistemlere aldığımız zaman kullanımı ücret karşılığında yapılabilir.
* Yani siteye girip aracının fiyatını tahmin edilmesini isteyen bir kişi bunu bir para karşılığında yapması yani karşılığında bir ücret ödemesi istenebilir.
